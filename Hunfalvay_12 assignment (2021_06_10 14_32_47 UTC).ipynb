```python
# DATA 620 Assignment 12
# Written by Melissa Hunfalvay
# Last Updated April 6th, 2021

# This program opens ﬁles Obama_2011.txt, Obama_2015.txt, Trump_2016-2020.txt, biden_2014.txt, Biden_2018.txt and reads it line by line. 
# Each line is then split into a list of words using the split function.
# Each word is then checked to see if the word is already in the dictonary. 
# If the word is not in the list, it is added to the dictionary.
# The words are sorted and printed by highest frequency to include only the top 30 most frequent words
# Stop words, symbols and characters are removed. Decisions are made for word stemming based on results.
 

import sys
import nltk
# Modified for UMUC DATA 620 by Paula Colliver and Nivedita Bijlani   
# Modified additionally for my own Week 12 project on cultural & legal changes by Melissa Hunfalvay on April 6th, 2021
import re
from nltk.corpus import stopwords # NLTK import of identified stop words
from nltk.stem import PorterStemmer # Stemmer takes the first part of the word - the word root e.g. Bush and Bushs and colates them into one count
from nltk.tokenize import word_tokenize
   
ps = PorterStemmer()

# Step 1.Download the text file to your local machine and for ease of use place this file in the same folder as the python file for this assignment. This is completed.

# Step 2:Create and initialize a dictionary for words 

d = {}

# Step 3: Prompt the user for the file name because tehre are multiple files to choose from. 
# Use a try-except block to exit with a user-friendly error message if there is an error opening the file name specified.

try:
    
    filename = input("Please enter a file name: ")  # Prompt the user for a file name
    if ".txt" not in filename: #Error Case 1: if the person adds the correct file name but does not add the txt extension
        filename = filename + ".txt"
    f = open(filename,"r") # Open a connection to the file, r = read only
    file = f.readlines() # Read each line of the file and store it to a variable, i.e. file
    whole_file = "".join(file) # Reads the entire file at once
    f.close() # This closes the file because it has been read
    # I am not lower casing the file name because upper and lower case files can legitimately exist and they would be seperate files
   
    # Remove stop words
    stop_words = stopwords.words('english')
    
    for line in file:
        words = line.split() # Step 5: Use the “split” function to split the line into individual words and collect these in a list. 
        if len(words)>0: # Check for the condition that the line may be empty, i.e. not contain any words!
            for w in words: 
                w = w.lower() # Error case 2: lower case each word so counting is uniform and accurate
                w = re.sub(r'[^\w\s]', '', w) # re is a library, the subfunction this will remove any pattern of punctuations and replace it with a blank argument 
                w = re.sub(r'[^\x00-\x7F]+','', w) # re is a library, the subfunction will remove all non- ascii nd non-alpha neumeric characters with a blank 
                w = re.sub("\d+", "", w) # re is a library, the subfunction will remove all numbers and replace with blank
                #w = ps.stem(w) # Take stem (root) of all the words, this will help with the count of words for instance constitution and constitutional
                # Note that there is a downside to word stemming as it is a global function, so make the decision to use this or not once you see the different file outputs.
                if w in d:    # Was the word previously seen
                    d[w] += 1
                else:
                    if (w not in stop_words) and (w != ""): # This says if teh word is not in teh word stop library and, second, that I will filter out blanks so they do not become part of the count
                        d[w] = 1

#count the words
    num_words = sum(d[w] for w in d)    
    
# count characters 
    num_chars = len(whole_file)

# count lines 
    num_lines = whole_file.count('\n')
    
    lst = [(d[w], w) for w in d]
    lst.sort()
    lst.reverse()

    print('Your input file has ' + str(num_chars) + ' number of characters.')
    print('Your input file has ' + str(num_lines) + ' number of lines.')
    print('Your input file has ' + str(num_words) + ' number of words.')


    i = 1
    print('\n The 30 most frequent words are \n')
    for count, word in lst[:30]:
        print('%2s.  %4s %s' % (i, count, word))
        i += 1
            
# Error case 3: If the user enters anything other than the filename detect their mistake using try and except and print an error message 
except Exception as err:
    print(err)
    print("Input not valid")

# End of script
```
